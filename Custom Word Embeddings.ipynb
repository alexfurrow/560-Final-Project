# Gathering the Relevant Columns

```python
filter = ["occasion", "style","color","embellishments","category"]

# drop unused columns
df = df.drop(["mpn", "created_at", "updated_at", "deleted_at", "brand_canonical_url", "labels", "Occasions", "Style", "bc_product_id", "product_color_id"], axis=1)

# filter attributes: occasion and style
df = df[df.attribute_name.isin(filter)]

df.head(3)
```

```python
import re
```

```python
input_df.head(1)
```

```python
df = input_df
```

```python
# Combine description and details columns
df["new_column"] = df["brand"] + " " + df["name"] + " " + df["description"] + " " + df["details"]

# Regex: keep the words starting with capital/small letters
df["new_column"] = df["description"].apply(str)
cleaned_column=[]
for item2 in df["new_column"]:
    item2 =  re.sub(r'[^a-zA-Z]'," ",item2).lower()
    cleaned_column.append(item2)
    
df["new_column"]=cleaned_column
df.head(3)
```

```python
# Tokenize the sentences into words
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer(r'\w+')
df["new_column"] = df["new_column"].apply(lambda x: tokenizer.tokenize(x))
df.head(3)
```

```python
# Remove stop words
from nltk.corpus import stopwords
stop = stopwords.words('english')
```

```python
cleaned = []
for item3 in df["new_column"]:
    new_words = []
    for word in item3:
        if word in stop:
            continue
        new_words.append(word)

    cleaned.append(new_words)
df["new_column"] = cleaned

df.head(3)
```

```python
# Lemmatize

from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

def word_lemmatizer(text):
    print(text)
    lem_text = [lemmatizer.lemmatize(i) for i in text]
    return lem_text

df["new_column"] = df["new_column"].apply(lambda x: word_lemmatizer(x))
```

# Custom Word Embeddings

```python
import spacy
```

```python
#custom embeddings
nlp = spacy.load("en_core_web_md")
```

```python
words = df['new_column']
```

```python
words = list(words)
```

```python
words[0]
```

```python
for i in range(0,len(words)):
    words[i] = " ".join(words[i])
```

```python
for idx, word in enumerate(words):
    print(nlp(word))
    print(nlp(word).vector[:10]) 
    
    if idx == 5: # stop printing after first 5 or so, takes a long time!
        break
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(words)

tf_idf_lookup_table = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names())
```

```python
DOCUMENT_SUM_COLUMN = "DOCUMENT_TF_IDF_SUM"

# sum the tf idf scores for each document
tf_idf_lookup_table[DOCUMENT_SUM_COLUMN] = tf_idf_lookup_table.sum(axis=1)
available_tf_idf_scores = tf_idf_lookup_table.columns # a list of all the columns we have
available_tf_idf_scores = list(map( lambda x: x.lower(), available_tf_idf_scores)) # lowercase everything
```

```python
import numpy as np

tweets_vectors = []
for idx, word in enumerate(words): # iterate through each review
    tokens = nlp(word) # have spacy tokenize the review text
    
    # initially start a running total of tf-idf scores for a document
    total_tf_idf_score_per_document = 0
    
    # start a running total of initially all zeroes (300 is picked since that is the word embedding size used by word2vec)
    running_total_word_embedding = np.zeros(300) 
    for token in tokens: # iterate through each token
    
    # if the token has a pretrained word embedding it also has a tf-idf score
        if token.has_vector and token.text.lower() in available_tf_idf_scores:
            
            tf_idf_score = tf_idf_lookup_table.loc[idx, token.text.lower()]
            #print(f"{token} has tf-idf score of {tf_idf_lookup_table.loc[idx, token.text.lower()]}")
            running_total_word_embedding += tf_idf_score * token.vector
            
            total_tf_idf_score_per_document += tf_idf_score
    
    # divide the total embedding by the total tf-idf score for each document
    document_embedding = running_total_word_embedding / total_tf_idf_score_per_document
    tweets_vectors.append(document_embedding)
```
